## annotation service

```python
import requests
```

```python
## 基于以下网址
baseurl = "http://myvariant.info/v1/variant/"
```

```python
## 查询网址格式
"http://myvariant.info/v1/variant/<variantid>"
```

```python
## example
response = requests.get("http://myvariant.info/v1/variant/chr1:g.35367G>A")
```

```python
response.encoding = response.apparent_encoding
```

```python
print response.text
```

```python
out = response.content
print out
```

### fileds
加入用户想找的信息，默认为=all

```python
print requests.get("http://myvariant.info/v1/variant/chr1:g.35367G>A?fields=cadd").text
```

```python
requests.get("http://myvariant.info/v1/variant/chr1:g.35367G>A?fields=exac").text
```

### 批量
逗号分割

```python
import httplib2
h = httplib2.Http()
headers = {'content-type': 'application/x-www-form-urlencoded'}
params = 'ids=chr16:g.28883241A>G,chr1:g.35367G>A&fields=dbnsfp.genename,cadd.gene'
res, con = h.request('http://myvariant.info/v1/variant', 'POST', params, headers=headers)
```
